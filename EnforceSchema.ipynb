```python
from pyspark.sql.types import StructType,IntegerType,StructField,FloatType,BooleanType,StringType
```

```python
!hadoop fs -head /data/first_100_customers.csv
```

## StructType

```python
schema=StructType([
    StructField('customer_id',IntegerType(),False),
    StructField('Name',StringType(),False),
    StructField('city',StringType(),False),
    StructField('state',StringType(),False),
    StructField('country',StringType(),False),
    StructField('registration_date',StringType(),False), 
    StructField('is_active',BooleanType(),False), 
])
```

```python
df=spark.read\
.format('csv')\
.option('header',"true")\
.schema(schema)\
.load('/data/first_100_customers.csv')
```

```python
df.show()
```

```python
df.printSchema()
```

## DDL Schema

```python
ddl_schema='customer_id INT, name BOOLEAN ,city STRING,state STRING,country STRING ,registration_data STRING ,is_active BOOLEAN'

## here name is in boolean type
```

```python
df_ddl=spark.read\
.format('csv')\
.option('header',"true")\
.schema(ddl_schema)\
.load('/data/first_100_customers.csv')
```

```python
df_ddl.printSchema()
```

```python
df_ddl.show()  ## name show all null because we provide wrong type
```
