```python
from pyspark.sql import SparkSession
#initialize spark session
spark=SparkSession.builder \
.appName("Repartition and Coalesce")\
.getOrCreate()
```

```python
!hadoop fs -head /user/msd24002/customers1mb.csv
```

```python
rdd=spark.sparkContext.textFile('/user/msd24002/customers1mb.csv')
```

```python
rdd.take(5)
```

```python
rdd.getNumPartitions()
```

```python
repartition_rdd=rdd.repartition(4)
```

```python
repartition_rdd.getNumPartitions()
```

```python
repartition_rdd_less=rdd.repartition(1)
```

```python
repartition_rdd_less.getNumPartitions()
```
